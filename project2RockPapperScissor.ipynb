```python
rock = "rock"
papper = "papper"
scissor = "scissor"
print("_"*85 + "\n")
while True:

    user1 = input("user1 please choose any one : ")
    user2 = input("user2 please choose any one : ")
    print("\n"+"*"*85)
    if(user1 == rock) and (user2 == rock):
        print("result: tie ..... please try again")
    elif(user1 == rock) and (user2 == papper):
        print("result: win user2")
    elif(user1 == rock) and (user2 == scissor):
        print("result: win user1")
    elif(user1 == papper) and (user2 == rock):
        print("result: win user1")
    elif(user1 == papper) and (user2 == papper):
        print("result: tie ..... please try again")
    elif(user1 == papper) and (user2 == scissor):
        print("result: win user2")
    elif(user1 == scissor) and (user2 == rock):
        print("result: win user2")
    elif(user1 == scissor) and (user2 == papper):
        print("result: win user1")
    elif(user1 == scissor) and (user2 == scissor):
        print("result: tie ..... please try again")
    else:
        break
    print("_"*85 + "\n")
print("result: Please choose any one out of rock, papper and scissor")
```
