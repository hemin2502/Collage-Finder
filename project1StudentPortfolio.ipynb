# Project 1 : Student portfolio

###  Please Enter Your Details

```python
studentName = str(input("Name: "))
studentStd = input("Standard: ")
studentDiv = input("Division: ")
studentRollNo = input("Roll no.: ")
```

### Please Enter Of Your Each Subjects Marks

```python
maths = int(input("Maths: "))
science = int(input("Science: "))
english = int(input("English: "))
history = int(input("History: "))
percentage = (maths + science + english + history)/4
```

### Student portfolio

```python
print("\n:::::::: Student Portfolio ::::::::\n")
print("Name: ",studentName)
print("Standard: ",studentStd)
print("Division: ",studentDiv)
print("RollNo.: ",studentRollNo)
print("\n:::::::: MARKS ENTRY ::::::::\n")
print("Maths: ",maths)
print("Science: ",science)
print("English: ",english)
print("History: ",history)
print("Net Percentage: ",percentage)
```
